```python
neuron_df, conn_df = fetch_adjacencies(NC(instance=KCs), NC(instance=DAN_a1+MBON_a1))

neuron_df = neuron_df[['bodyId', 'instance']]
neuron_df
# conn_df
```

```python
from tqdm import tqdm
import time

for i in tqdm(range(0, 5)):
    time.sleep(2)
```

```python
m = set()
n = set()

for i in range(len(instance_from_neurons)):
    m.update(set(neuron_df.loc[neuron_df.instance == instance_from_neurons[i]].bodyId))

for i in range(len(instance_to_neurons)):
    n.update(set(neuron_df.loc[(neuron_df.instance == instance_to_neurons[i])].bodyId))

print("from neuron ids:", m)
print("to neuron ids:", n)
```

```python
# remove connections lesser than 'x' number of synapses
total = 0
drop = []

for ind in conn_df.index:
    total +=1
    if (conn_df['bodyId_pre'][ind] in m) or (conn_df['bodyId_post'][ind] in m):
        
        if conn_df['weight'][ind] < 1:
            drop.append(ind)
            
    else:
        if(conn_df['weight'][ind] < 1):
            drop.append(ind)
        
print("all edges:", total)
print("remaining edges:", total - len(drop))

conn_df.head()
```

```python
adj_mat = np.zeros((len(m), len(n)))
pre_dict = {}
post_dict = {}
pre_i, post_j = (0, 0)

for ind in conn_df.index:
    if conn_df['bodyId_pre'][ind] in pre_dict:
        i = pre_dict[conn_df['bodyId_pre'][ind]]
    else:
        i = pre_i
        pre_i += 1
        pre_dict[conn_df['bodyId_pre'][ind]] = i
    
    if conn_df['bodyId_post'][ind] in post_dict:
        j = post_dict[conn_df['bodyId_post'][ind]]
    else:
        j = post_j
        post_j += 1
        post_dict[conn_df['bodyId_post'][ind]] = j

    adj_mat[i][j] += conn_df['weight'][ind]
```

```python
plt.rcParams["figure.figsize"] = (2, 10)
p = sns.heatmap(adj_mat, cmap="YlGnBu", square=True)
p.set_xlabel(instance_to_neurons)
p.set_ylabel(instance_from_neurons)
plt.savefig("connection_vis_KC.png", dpi=700)
```

```python
# normalize connection weights + create netx graph
# conn_df.weight /= conn_df['weight'].max(axis=0)

G = nx.from_pandas_edgelist(conn_df, source='bodyId_pre', target='bodyId_post', edge_attr=['weight'], create_using=nx.DiGraph())
print("#nodes: ", G.number_of_nodes())
print("#edges: ", G.number_of_edges())

print("connectivity df:")
print(conn_df.head())
```

```python
nx.set_node_attributes(G, dict(neuron_df.to_dict(orient="tight")['data']), name="instance")
# G = nx.convert_node_labels_to_integers(G)
G.nodes.data()
```

```python
plt.figure(figsize=(7,7))
# nx.draw_networkx_edges(G, pos, alpha=0.3, width=edgewidth, edge_color="m")
# nx.draw_networkx_nodes(G, pos, alpha=0.9)

def color_instances(node_attrs):
    if node_attrs['instance'] == 'MBON07(a1)_R':
        return {'color': 'red'}
    elif node_attrs['instance'] == 'PAM11(a1)_R':
        return {'color': 'yellow'}
    else:
        return {'color': 'black'}

# fig, ax = plt.subplots()
plot_network(G, layout='circular', node_style=color_instances)

plt.savefig("netx_graph_kcab-p.png", dpi=700) 
plt.show()
```
